# Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.decomposition import PCA
import sklearn as sk
```

# Read in data

```python
df = pd.read_csv('/Users/zoebjamal/Desktop/python/mlgenetics_hw7_TRAIN.txt', sep=r'\s{1,}', header = None, engine = 'python')
df
```

```python
batch = pd.read_csv('/Users/zoebjamal/Desktop/python/mlgenetics_hw7_TRAIN_batch.txt', sep=r'\s{1,}', header = None, engine = 'python', names = ['batch'])
batch
```

# Scale and center the data

```python
scaled = sk.preprocessing.scale(df)
```

# Run PCA

```python
pca = PCA(n_components = 10)
```

```python
pcs = pca.fit_transform(scaled)
```

# 3a)

```python
a = pca.explained_variance_ratio_
```

```python
a.sum()
```

# 3b)

```python
pcs = pd.DataFrame(data = pcs)
pcs.columns = pcs.columns.astype(str)
```

```python
# add batch as a column
pcs['batch'] = batch
pcs
```

```python
# find correlation matrix - look at 
corr = pcs.corr(method = 'pearson')
corr['batch'] = corr['batch'] ** 2
corr
```

# 3c)

```python
comps = pd.DataFrame(data = pca.components_) # get loadings
comps.columns = comps.columns.astype(str) # convert column names to strings
comps = comps.abs() # get absolute values
# this matrix has 10 rows and 5000 columns
```

```python
maxes = comps.max().to_frame() # find the max value per column
maxes.columns = maxes.columns.astype(str)
# the maxes matrix has 5000 rows (the max value for each site per PC loading)
```

```python
top1k = maxes.nlargest(1000, '0') # get the top 1000 sites
```

```python
newSites = list(top1k.index.values) # convert the index to a list so we know which sites to take
```

```python
df = df.T # transpose data so we can use the indices we just got
```

```python
sites = df.iloc[newSites] # subset according to the top 1000 sites
sites = sites.T # transpose back so we can run PCA
```

```python
sites
```

```python
scaled2 = sk.preprocessing.scale(sites) # center and scale new data again
pca2 = PCA(n_components = 10)
pcs2 = pca2.fit_transform(scaled2)
```

```python
pcs2 = pd.DataFrame(data = pcs2)
pcs2.columns = pcs2.columns.astype(str)
```

```python
pcs2['batch'] = batch
```

```python
corr2 = pcs2.corr(method = 'pearson')
corr2['batch'] = corr2['batch'] ** 2
corr2
```
