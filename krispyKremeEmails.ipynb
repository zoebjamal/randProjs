# Import Data and Format

```python
df = pd.read_csv('/Users/zoebjamal/Desktop/python/randprojs/KRISPYKREME_W22.csv', index_col = 0)
df.drop(['Name', 'Quantity', 'Unnamed: 5'], inplace = True, axis = 1)
df.rename(columns = {'Link' : 'Name', 'Unnamed: 6': 'link'}, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
class Person:
    """
    Class to store a Person's email, name, and a list of links that they are owed
    """
    def __init__(self, email, name, links):
        self.name = name
        self.email = email
        self.links = links
```

```python
# temp df needed to get unique people objects
temp = df.drop_duplicates(subset = 'Email',)
temp = temp.reset_index(drop = True)
```

```python
# create list of person objects
people = []
for i in range(len(temp)):
    newPerson = Person(temp.iloc[i]['Email'], temp.iloc[i]['Name'], [])
    people.append(newPerson)
```

```python
# add links to person if email matches
for i in range(len(people)):
    for j in range(len(df)):
        if people[i].email == df.iloc[j]['Email']:
            people[i].links.append(df.iloc[j]['link'])
```
